A smart travel guide for China. Fill in your profile and get personalised destination recommendations, maps, climate analysis and more.

```python
#Profile

from WorldExplorer.traveller_profile import traveller_profile

profile = traveller_profile()
```

```python
#Match Destinations

from WorldExplorer.match_destinations import match_destinations

df = match_destinations(profile)
df
```

```python
# Attraction map

from WorldExplorer.attractions_map import attractions_map

top5_regions = df["region"].tolist()
print(f"showing map for: {top5_regions}")

m = attractions_map(profile=profile, regions=top5_regions)
m
```

```python
# Seasonal analysis

from WorldExplorer.seasonal_analysis import seasonal_analysis
from IPython.display import Image, display

for region_name in top5_regions:
    print(f"\n--- {region_name} ---")
    seasonal_analysis(region_name, profile=profile)
    img_path = f"data/seasonal_{region_name.lower().replace(' ', '_')}.png"
    display(Image(img_path))
```

```python
#DashBoard Export

from WorldExplorer.country_dashboard_data import country_dashboard_data

data = country_dashboard_data(profile=profile)

print("\nstats for your profile:")
print(data["stats"][["region", "attraction_count", "avg_rating", "avg_cost_night", "budget_level"]])

print("\nmust-sees and hidden gems:")
print(data["mustsees"][["region", "name", "type"]])
```

## What was exported
All data has been saved to the `data/` folder:
- `top5_destinations.csv` — your personalised top 5
- `current_profile.csv` — your travel profile
- `dashboard_attractions.csv` — filtered attractions
- `dashboard_budget.csv` — budget info per region
- `dashboard_mustsees.csv` — must-sees and hidden gems
- `dashboard_stats.csv` — stats per region
- `worldexplorer.db` — SQLite database for Node-RED
- `attractions_map.html` — interactive map
